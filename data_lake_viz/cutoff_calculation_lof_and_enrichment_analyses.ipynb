# Visualize Data and Calculate Cutoffs
#### Calculate ideal cutoff value from combined dataset. cutoff value is point that maximizes (discovery/total gene list size), where total gene list size > N (default 20) genes

## LOF_Z

```r
ggplot(lof_z_grouped, aes(greater_or_equal, n_2018, fill = has_omim_phenotype)) +
            #geom_bar(stat="identity", position = "dodge") +
            geom_line(aes(color = has_omim_phenotype)) +
            #scale_fill_brewer(palette = "Reds", direction = -1) +
            scale_fill_brewer(palette = "Paired") +
            ylab("# Genes") +
            scale_y_log10() +
            xlab(paste("LoF Intolerance Z-score", " ≥")) +
            scale_x_continuous(breaks=seq(0,10,1), limits=c(-0.01, 10)) +
            ggtitle("")  +
            theme_Publication()
ggplot(lof_z_grouped, aes(greater_or_equal, n_2018, fill = has_omim_phenotype)) +
            #geom_bar(stat="identity", position = "dodge") +
            geom_line(aes(color = has_omim_phenotype)) +
            #scale_fill_brewer(palette = "Reds", direction = -1) +
            scale_fill_brewer(palette = "Paired") +
            ylab("# Genes") +
            #scale_y_log10() +
            scale_y_continuous(limits=c(0,150)) +
            xlab(paste("LoF Intolerance Z-score", " ≥")) +
            scale_x_continuous(breaks=seq(0,10,1), limits=c(6.5, 10)) +
            ggtitle("") +
            theme_Publication()
```

```r
lof_z_cutoff = calculate_cutoff_filtered(lof_z_grouped)
lof_z_cutoff
multiplot(get_cutoff_violin_plot(lof_z_grouped, N_SIMS, xlabel="Combined", ylabel="LoF Intolerance Z-score Cutoff Optimizing DD"), get_cutoff_violin_plot(lof_z_grouped_bgl, N_SIMS, xlabel="BG"), get_cutoff_violin_plot(lof_z_grouped_cmg, N_SIMS, xlabel="CMG"), cols=3)
# Calculate cutoff distribution
lof_z_ecdf = ecdf(get_sim_cutoffs(lof_z_grouped, N_SIMS))
lof_z_ecdf(lof_z_cutoff$value - 0.0001)
```

```r
ggplot(lof_z_grouped, aes(x=greater_or_equal, y=discovery_density_2013to2018)) + geom_line() + ylab("Discovery Density 2013-2018") + xlab("LoF Intolerance Z-score ≥") + geom_vline(xintercept=lof_z_cutoff$value, linetype='dotted') + theme_Publication()

plot_dd_year_combinations(lof_z_grouped, "LoF Z ≥")
```

### Check for overlap between newest OMIM (May 2019) gene list and candidate gene list

```r
omim_disease_newest = read.table("omim/omim_firstkeyonly_05_30_2019_noheader.txt", header=FALSE, sep="\t", quote = "", comment.char = "")
colnames(omim_disease_newest) = c("phenotype", "all_genes", "omim_id", "cyto_band", "gene_id")
omim_disease_newest = omim_disease_newest %>% group_by(gene_id) %>% summarise(phenos = toString(phenotype)) %>% ungroup()
omim_disease_newest$gene_id = trimws(as.character(omim_disease_newest$gene_id))

omim_disease = read.table("omim/omim_firstkeyonly_01_08_2018_noheader.txt", header=FALSE, sep="\t", quote= "", comment.char = "")
colnames(omim_disease) = c("phenotype", "all_genes", "omim_id", "cyto_band", "gene_id")
omim_disease = omim_disease %>% group_by(gene_id) %>% summarise(phenos = toString(phenotype)) %>% ungroup()
omim_disease$gene_id = trimws(as.character(omim_disease$gene_id))
```

```r
all_possible_genes = as.list(trimws(as.character(read.table("possible_genes/hg19_refGene.inCapture.name", header = FALSE)$V1)))
all_possible_genes_non_omimJan = setdiff(all_possible_genes, omim_disease$gene_id)
```

```r
omim_newest_non_omim_jan2018 = setdiff(omim_disease_newest$gene_id, omim_disease$gene_id)
length(omim_newest_non_omim_jan2018)
intersect(omim_newest_non_omim_jan2018, trimws(all_candidates))
# Previously, analysis was run when newest OMIM data was from Nov 12 2018. Then, length(omim_newest_non_omim_jan2018)=535, w/ 11 genes at the intersect
# 11 'new' genes not at intersect: ATP1A1, CLTC, FBXO11, IRF2BPL, KDM5B, KIAA1109, LINGO1, PDE1C, SCN3A, SET, VPS13D
# Feb 25, 2019, 222, – 18 genes – above plus CACNA1E, CHD3, DOCK3, MACF1, MAST1, MYO9A, TBX2
# May 30, 2019, 291, – 19 genes – above plus TCF20
```

### For candidate disease gene list, calculate enrichment with newest OMIM (Feb 2019) disease genes vs null distribution of random non-OMIM genes

```r
intersect_sizes = vector("numeric", 1)
for (i in 1:100000){
    cur_sample = as.character(unlist(sample(all_possible_genes_non_omimJan, size = length(all_candidates), replace=FALSE)))
    cur_intersect = intersect(cur_sample, omim_newest_non_omim_jan2018)
    intersect_sizes[i] = length(cur_intersect)
}
get_p_val = ecdf(intersect_sizes)
get_p_val(19)
# Nov 2018 -> Result of get_p_val(11) from first run (@10k sims) -> "1"
# Feb 2019 -> Result of get_p_val(18) from first run (@10k sims) -> "1"
# May 2019 -> Result of get_p_val(19) from first run (@100k sims) -> "1"
```

```r
mean(intersect_sizes) # Nov 2018 mean from 1st run -> 1.2834 | # Feb 2019 mean from 1st run -> 1.7807 | May 2019 mean from 1st run -> 2.28512
hist(intersect_sizes)
```
